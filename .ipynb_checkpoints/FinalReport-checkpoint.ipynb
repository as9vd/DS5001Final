## Latent Dirichlet Allocation Topic Modeling

Latent Dirichlet Allocation analyses and discovers themes collections of documents. In this case, I'll be using it to try and find themes across the 11 books, even though they are all different from one another. I will then use it on books for each era of midfielder.

```python
ngram_range = (1,2)
n_terms = 4000
n_topics = 40
max_iter = 20
n_top_terms = 9
```

```python
OHCO = ['book_id','chap_num','para_num','sent_num','token_num']
BAG = OHCO[:2]
CORPUS = pd.read_csv("TOKENS.csv").set_index(["book_id", "chap_num", "para_num", "sent_num", "token_num"])
CORPUS = CORPUS.dropna(subset=['term_str'])
```

```python
with warnings.catch_warnings():
    warnings.filterwarnings("ignore")
    
    noun_pattern = re.compile(r'^NN|^NNS|^NNP|^NNPS')
    DOCS = TOKENS[TOKENS['pos'].str.match(noun_pattern)]
    DOCS['term_str'] = DOCS['term_str'].astype(str)

    DOCS = (DOCS.groupby(['book_id', 'chap_num'])['term_str']
            .apply(lambda x: ' '.join(x)).to_frame().rename(columns={'term_str': 'doc_str'}))
    
    DOCS['doc_str'] = DOCS['doc_str'].str.replace('nan', '', case=False)
```

This is the DOCS table. We keep the nouns and turn everything into lowercase and strip extra spaces/symbols so we can feed it into the SKLearn library.

```python
count_engine = CountVectorizer(max_features=n_terms, ngram_range=ngram_range, stop_words='english')
count_model = count_engine.fit_transform(DOCS.doc_str)
TERMS = count_engine.get_feature_names_out()

VOCAB = pd.DataFrame(index=TERMS)
VOCAB.index.name = 'term_str'
DTM = pd.DataFrame(count_model.toarray(), index=DOCS.index, columns=TERMS)
```

```python
VOCAB['doc_count'] = DTM.astype('bool').astype('int').sum()
DOCS['term_count'] = DTM.sum(1)
lda_engine = LDA(n_components=n_topics, max_iter=max_iter, learning_offset=50., random_state=0)
TNAMES = [f"T{str(x).zfill(len(str(n_topics)))}" for x in range(n_topics)]
lda_model = lda_engine.fit_transform(count_model)
```

#### THETA, PHI, TOPICS

THETA is the document-topic distribution. Each row contains the probability of each topic in each document within the corpus. It helps distinguish the the dominant topics for each chapter in my case.

PHI is the topic-word distribution. Each row contains the probability of each word belonging to each topic in the model. It helps distinguish the dominant words for each topic within the corpus.

I ran these on individual eras. But first, the results below are for the entire corpus. As you can see, the Ferguson books by Taylor and Barclay are closely related (for obvious reasons). Interestingly enough, though, it is evident that Robson and Keane are closely affiliated with one another in the dendrogram, as are Scholes and Charlton. This is an interesting observation, as Robson and Keane were two fantastic leaders and captains. Scholes and Charlton were also leaders, but it's not what they were known for, whereas Robson and Keane were known to be loud and leaders of men. 

In the end, it seems that scikit-learn's LDA vectoriser can point out obvious patterns in text, but it's not the level of depth I was hoping for. This is good as a bit of a sanity check, but I need to get deeper. Next section will look at sentiment, and hopefully that'll scratch my itch. 

```python
def produce_dendrogram(TOKENS, book_ids, authors, ngram_range = (1,2), n_terms = 4000, n_topics = 20, max_iter = 20, n_top_terms = 15):
    OHCO = ['chap_num','para_num','sent_num','token_num']
    BAG = OHCO[1:2]
    CORPUS = pd.read_csv("TOKENS.csv").set_index(["book_id", "chap_num", "para_num", "sent_num", "token_num"])
    CORPUS = CORPUS.loc[[2, 4, 5]]
    CORPUS = CORPUS.dropna(subset=['term_str'])

    with warnings.catch_warnings():
        warnings.filterwarnings("ignore")
    
        noun_pattern = re.compile(r'^NN|^NNS|^NNP|^NNPS')
        TOKENS = TOKENS.loc[book_ids]
        DOCS = TOKENS[TOKENS['pos'].str.match(noun_pattern)]
        DOCS['term_str'] = DOCS['term_str'].astype(str)

        DOCS = (DOCS.groupby(['book_id', 'chap_num'])['term_str']
                    .apply(lambda x: ' '.join(x)).to_frame().rename(columns={'term_str': 'doc_str'}))
    
        DOCS['doc_str'] = DOCS['doc_str'].str.replace('nan', '', case=False)
    
        count_engine = CountVectorizer(max_features=n_terms, ngram_range=ngram_range, stop_words='english')
        count_model = count_engine.fit_transform(DOCS.doc_str)
        TERMS = count_engine.get_feature_names_out()

        VOCAB = pd.DataFrame(index=TERMS)
        VOCAB.index.name = 'term_str'
        DTM = pd.DataFrame(count_model.toarray(), index=DOCS.index, columns=TERMS)
    
        VOCAB['doc_count'] = DTM.astype('bool').astype('int').sum()
        DOCS['term_count'] = DTM.sum(1)
        lda_engine = LDA(n_components=n_topics, max_iter=max_iter, learning_offset=50., random_state=0)
        TNAMES = [f"T{str(x).zfill(len(str(n_topics)))}" for x in range(n_topics)]
        lda_model = lda_engine.fit_transform(count_model)
        
    THETA = pd.DataFrame(lda_model, index=DOCS.index)
    THETA.columns.name = 'topic_id'
    THETA.columns = TNAMES

    PHI = pd.DataFrame(lda_engine.components_, columns=TERMS, index=TNAMES)
    PHI.index.name = 'topic_id'
    PHI.columns.name  = 'term_str'

    TOPICS = PHI.stack().to_frame('topic_weight').groupby('topic_id')\
                .apply(lambda x: x.sort_values('topic_weight', ascending=False)\
                .head(n_top_terms).reset_index().drop('topic_id', axis=1)['term_str'])

    TOPICS['label'] = TOPICS.apply(lambda x: x.name + ' ' + ', '.join(x[:n_top_terms]), 1)
    TOPICS['doc_weight_sum'] = THETA.sum()
    TOPICS['term_freq'] = PHI.sum(1) / PHI.sum(1).sum()
    
    LIB['author_key'] = LIB.author.str.split(' ').str[-1].str.lower()

    AUTHORS = authors

    TOPICS[AUTHORS] = THETA.join(LIB, on='book_id').groupby('author_key')[TNAMES].mean().T
    TOPICS['author'] = TOPICS[AUTHORS].idxmax(1)
    
    return PHI, THETA, TOPICS
```

```python
PHI, THETA, TOPICS = produce_dendrogram(TOKENS, [i for i in range(0, 12) if i != 3])
TOPICS = TOPICS.drop(["T02", "T08", "T37", "T36", "T35", "T34", "T31", "T29", "T21", "T17"])
PHI = PHI.drop(["T02", "T08", "T37", "T36", "T35", "T34", "T31", "T29", "T21", "T17"])
THETA = THETA.T
THETA = THETA.drop(["T02", "T08", "T37", "T36", "T35", "T34", "T31", "T29", "T21", "T17"])
```

```python
import scipy.cluster.hierarchy as sch
from scipy.spatial.distance import pdist #, squareform
# from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
import matplotlib.pyplot as plt            

class HAC:
    """This class takes an arbitrary vector space and represents it 
    as a hierarhical agglomerative cluster tree. The number of observations
    should be sufficiently small to allow being plotted."""

    w:int = 10
    labelsize:int = 14
    orientation:str = 'left'
    sim_metric:str = 'cosine' # The distance metric to use.
    tree_method:str = 'ward'
    norm:str = 'l2' # l1, l2, max
    
    def __init__(self, M, labels=None):
        self.M = M
        self.h = M.shape[0]
        if labels:
            self.labels = labels            
        else:
            self.labels = M.index.tolist()

    def get_sims(self):
        self.SIMS = pdist(normalize(self.M, norm=self.norm), metric=self.sim_metric)

    def get_tree(self):
        self.TREE = sch.linkage(self.SIMS, method=self.tree_method)        
        
    def plot_tree(self):
        plt.figure()
        fig, axes = plt.subplots(figsize=(self.w, self.h / 3))
        dendrogram = sch.dendrogram(self.TREE, labels=self.labels, orientation=self.orientation);
        plt.tick_params(axis='both', which='major', labelsize=self.labelsize)
        
    def plot(self):
        self.get_sims()
        self.get_tree()
        self.plot_tree()
```

This is what it looks like for the overall corpus. Not too many patterns here, which is to be expected, since the books cover events from 1909 to 2013. But it is very interesting to see. Atletico Madrid, Barcelona, and Real Madrid all get their own topic in topic 30, whilst the Robin Van Persie transfer saga in 2012 gets its own topic in topic 22. Very interesting stuff.

```python
labels  = [f"{a.upper()} {b}" for a, b in zip(TOPICS.author, TOPICS.label)]
HAC(PHI, labels=labels).plot();
```
